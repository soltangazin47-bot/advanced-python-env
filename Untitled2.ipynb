```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Score": [85, 78, 92, 88, 95]
}

df = pd.DataFrame(data)
df
```

```python
plt.plot(df["Name"], df["Score"])
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Line Plot")
plt.show()
```
